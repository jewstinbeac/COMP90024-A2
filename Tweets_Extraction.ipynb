```python
from collections import defaultdict

def tweet_reply_map(tweet_ids_file_path, labels_file_path):
    tweet_ids = []
    labels = []
    
    with open(tweet_ids_file_path) as file:
        for line in file:
            tweet_ids.append(line.strip().split(","))
            
    with open(labels_file_path) as file:
        for line in file:
            labels.append(line.strip())
            
    return tweet_ids, labels
```

```python
import os
import json
tweets_not_found = set()

def extract_tweets(source_file_path, target_dir):
    
    # extract tweet-ids from a txt file and save them in a dict
    tweet_ids = set()
    with open(source_file_path) as file:
        for line in file:
            tweet_ids.update(line.strip().split(","))
            
    tweet_ids = list(tweet_ids)        
    
    # extract tweets from twitter using the tweet-ids
    files = os.listdir(target_dir)
    print(files)
    unique_tweet_ids = set([file.replace('.json', '') for file in files])
            
    for i in range(0, len(tweet_ids), 100):
        
        print(f'Round {i}')
        
        end = i + 100
        if end > len(tweet_ids):
            end = len(tweet_ids)
        
        lookup_arr = [tweet_ids[j] for j in range(i, end) if tweet_ids[j] not in unique_tweet_ids]
        
        try:
            tweets = api.lookup_statuses(lookup_arr, map=True)
            print(len(tweets))
            for tweet in tweets:
#                 print(len(tweet._json))
#                 print(f"{'created_at' in tweet._json, tweet._json['created_at']}")
                if 'created_at' in tweet._json:
                    unique_tweet_ids.add(tweet._json['id_str'])
                    # save the source tweet in a json file
                    with open(f"{target_dir}/{tweet._json['id_str']}.json", 'w') as file:
#                         print(type(tweet._json))
                        json.dump(tweet._json, file)
                else:
                    print(f"{tweet._json['id']} not found")
                    tweets_not_found.add(tweet._json['id'])
                
        except Exception as exc:
            print(exc)
            
```

```python
extract_tweets(source_file_path = './train.data.txt', target_dir = './train_tweets')
```
